```python
import pandas as pd

# Loading Data
df = pd.read_csv("/content/ToyotaCorolla - MLR.csv")  # Already loaded earlier

# Preview data and cleanigg
print(df.head())           # First 5 rows
print(df.shape)            # Number of rows and columns
print(df.columns)          # List of column names
```

```python
#Checking Data types
print(df.info())

# Checking missing values
print(df.isnull().sum())
```

```python
# Numerical Values
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ploting histogram for each numerical column for more detail
df.hist(bins=30, figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
print(df.columns)
```

```python
df.rename(columns={'Fuel_Type': 'FuelType'}, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='FuelType', data=df)
plt.title("Fuel Type Distribution")
plt.show()
```

```python
print(df.columns.tolist())
```

```python
# Number of each fuel type
sns.countplot(x='FuelType', data=df)
plt.title("Fuel Type Distribution")
plt.show()
```

```python
df.rename(columns={'Age_08_04': 'Age'}, inplace=True)
```

```python
sns.scatterplot(x='Age', y='Price', data=df)
```

```python
# Relation between Age and Price
sns.scatterplot(x='Age', y='Price', data=df)
plt.title("Price vs Age")
plt.show()

# KM vs Price
sns.scatterplot(x='KM', y='Price', data=df)
plt.title("Price vs KM")
plt.show()
```

```python
sns.boxplot(x='FuelType', y='Price', data=df)
plt.title("Price by Fuel Type")
plt.show()
```

```python
# Convert FuelType
df_encoded = pd.get_dummies(df, columns=['FuelType'], drop_first=True)

# Correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(df_encoded.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Boxplots for numerical features
numerical_cols = ['Price', 'Age', 'KM', 'HP', 'cc', 'Weight']
for col in numerical_cols:
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot for {col}")
    plt.show()
```

```python
# Measure
print(df[numerical_cols].skew())

print(df[numerical_cols].kurtosis())
```
